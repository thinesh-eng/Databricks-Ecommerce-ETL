```python
from pyspark.sql.functions import *
```

```python
orders_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv("/Volumes/workspace/default/bronze/orders.csv")
)
```

```python
orders_df.show(5)
```

```python
orders_df.printSchema()
```

```python
print("Total Records :", orders_df.count())
```

```python
orders_df.describe().show()
```

```python
orders_df.filter(
    col("order_id").isNull() |
    col("customer_id").isNull() |
    col("product_id").isNull()
).show()
```

```python
orders_df.groupBy("order_id") \
.count() \
.filter(col("count") > 1) \
.show()
```

```python
orders_df.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("workspace.default.bronze_orders")
```
