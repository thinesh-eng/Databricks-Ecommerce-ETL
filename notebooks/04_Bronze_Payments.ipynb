```python
payments_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv("/Volumes/workspace/default/bronze/payments.csv")
)
```

```python
payments_df.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("workspace.default.bronze_payments")
```
