```python
#Production_Style

from datetime import datetime
import traceback

table_name = "silver_customer_test"

try:

    df = spark.table(table_name)

    count = df.count()

    status = "SUCCESS"

    error_message = None

except Exception as e:

    status = "FAILED"

    error_message = str(e)

print("Status :", status)

print("Error :", error_message)
```

```python
from pyspark.sql import Row
from datetime import datetime

log_df = spark.createDataFrame([
    Row(
        run_id="RUN_002",
        pipeline_name="Ecommerce_ETL_Pipeline",
        task_name="Validation_Framework",
        start_time=datetime.now(),
        end_time=datetime.now(),
        status=status,
        records_processed=0,
        error_message=error_message
    )
])

log_df.write.mode("append").saveAsTable("pipeline_log")

print("Pipeline log inserted successfully.")
```

```python
%sql
SELECT *
FROM pipeline_log
ORDER BY start_time DESC;
```
