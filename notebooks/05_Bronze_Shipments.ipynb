```python
shipments_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv("/Volumes/workspace/default/bronze/shipments.csv")
)
```

```python
shipments_df.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("workspace.default.bronze_shipments")
```
