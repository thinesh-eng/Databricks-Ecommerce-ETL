```python
customers_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv("/Volumes/workspace/default/bronze/customers.csv")
)

customers_df.show(5)
```

```python
customers_df.printSchema()
```

```python
customers_df.count()
```

```python
customers_df.describe().show()
```

```python
from pyspark.sql.functions import col

customers_df.filter(
    col("customer_id").isNull()
).show()
```

```python
customers_df.groupBy("customer_id") \
.count() \
.filter("count>1") \
.show()
```

```python
customers_df.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("workspace.default.bronze_customers")
```

```python
spark.sql("""
SELECT *
FROM workspace.default.bronze_customers
LIMIT 10
""").show()
```
