```python
from pyspark.sql.functions import *
```

```python
products_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv("/Volumes/workspace/default/bronze/products.csv")
)
```

```python
products_df.show(5)
```

```python
products_df.printSchema()
```

```python
print("Total Records :", products_df.count())
```

```python
products_df.describe().show()
```

```python
products_df.filter(
    col("product_id").isNull() |
    col("product_name").isNull() |
    col("price").isNull()
).show()
```

```python
products_df.groupBy("product_id") \
.count() \
.filter(col("count") > 1) \
.show()
```

```python
products_df.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("workspace.default.bronze_products")
```

```python
spark.sql("""
SELECT *
FROM workspace.default.bronze_products
LIMIT 10
""").show()
```
